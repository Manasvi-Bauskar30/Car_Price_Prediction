# CAR PRICE PREDICTION - EDA

```python
# 1Ô∏è Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load Dataset
data = pd.read_csv("data/car_price_dataset.csv")

print("Dataset Loaded Successfully!\n")
```

```python
# Basic Information

print("Shape of dataset:", data.shape)
print("\nColumns:\n", data.columns)
print("\nData Types:\n")
print(data.dtypes)
```

```python
# First 5 Rows

print("\nFirst 5 Rows:\n")
print(data.head())
```

```python
# Missing Values

print("\nMissing Values:\n")
print(data.isnull().sum())
```

```python
# Statistical Summary

print("\nStatistical Summary:\n")
print(data.describe())
```

## VISUALIZATION SECTION

```python
# Price Distribution

plt.figure(figsize=(6,4))
sns.histplot(data['Price_INR'], kde=True)
plt.title("Price Distribution")
plt.xlabel("Price (INR)")
plt.ylabel("Count")
plt.show()
```

```python
# Brand vs Price

plt.figure(figsize=(10,5))
sns.boxplot(x='Brand', y='Price_INR', data=data)
plt.xticks(rotation=45)
plt.title("Brand vs Price")
plt.show()

```

```python
# Age vs Price

plt.figure(figsize=(6,4))
sns.scatterplot(x='Age_Years', y='Price_INR', data=data)
plt.title("Car Age vs Price")
plt.xlabel("Car Age (Years)")
plt.ylabel("Price (INR)")
plt.show()
```

```python
 # Mileage vs Price

plt.figure(figsize=(6,4))
sns.scatterplot(x='Mileage_KM', y='Price_INR', data=data)
plt.title("Mileage vs Price")
plt.xlabel("Mileage (KM)")
plt.ylabel("Price (INR)")
plt.show()
```

```python
# Power vs Price

plt.figure(figsize=(6,4))
sns.scatterplot(x='Power_BHP', y='Price_INR', data=data)
plt.title("Power vs Price")
plt.xlabel("Power (BHP)")
plt.ylabel("Price (INR)")
plt.show()
```

```python
# Correlation Heatmap (Only numeric columns) 

numeric_data = data.select_dtypes(include=['int64','float64'])

plt.figure(figsize=(8,6))
sns.heatmap(numeric_data.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

print("\nEDA Completed Successfully!")
```
